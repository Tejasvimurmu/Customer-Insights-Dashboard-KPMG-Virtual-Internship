```python
import pandas as pd

# Correct file name
file_path = 'KPMG_VI_New_raw_data_update_final.xlsx'

# Load sheets
customer_demo = pd.read_excel(file_path, sheet_name='CustomerDemographic')
customer_address = pd.read_excel(file_path, sheet_name='CustomerAddress')
transactions = pd.read_excel(file_path, sheet_name='Transactions')
new_customer = pd.read_excel(file_path, sheet_name='NewCustomerList')

# Display top rows
print("CustomerDemographic")
display(customer_demo.head())

print("CustomerAddress")
display(customer_address.head())

print("Transactions")
display(transactions.head())

print("NewCustomerList")
display(new_customer.head())
```

```python
#skip the not rows and load the real data
customer_demo = pd.read_excel(file_path, sheet_name='CustomerDemographic', skiprows=1)
customer_demo.columns = customer_demo.columns.str.strip()
print(customer_demo.columns.tolist())

# Check info and nulls
customer_demo.info()
customer_demo.isnull().sum()

# Clean: Drop rows with missing DOB or Gender
customer_demo = customer_demo.dropna(subset=['DOB', 'gender'])

#Convert 'DOB' to datetime
customer_demo['DOB'] = pd.to_datetime(customer_demo['DOB'], errors='coerce')

# Convert 'DOB' to age
from datetime import datetime
customer_demo['age'] = customer_demo['DOB'].apply(lambda x: datetime.now().year - x.year)

# Normalize gender values
customer_demo['gender'] = customer_demo['gender'].replace({'F': 'Female', 'M': 'Male', 'U': 'Unknown'})
```

```python
#skip the not rows and load the real data
transactions = pd.read_excel(file_path, sheet_name='Transactions', skiprows=1)
transactions.columns = transactions.columns.str.strip()
print(transactions.columns.tolist())

# Check nulls and data types
transactions.info()
transactions.isnull().sum()

# Drop nulls
transactions = transactions.dropna(subset=['product_id', 'transaction_date', 'online_order'])

# Convert dates
transactions['transaction_date'] = pd.to_datetime(transactions['transaction_date'])
```

```python
#skip the not rows and load the real data
customer_address = pd.read_excel(file_path, sheet_name='CustomerAddress', skiprows=1)
customer_address.columns = customer_address.columns.str.strip()
print(customer_address.columns.tolist())

# Basic check
customer_address.info()
customer_address.isnull().sum()

# Drop rows with null customer_id
customer_address = customer_address.dropna(subset=['customer_id'])
```

```python
#skip the not rows and load the real data
new_customer = pd.read_excel(file_path, sheet_name='NewCustomerList', skiprows=1)
new_customer.columns = new_customer.columns.str.strip()
print(new_customer.columns.tolist())

# Check structure
new_customer.info()
new_customer.isnull().sum()

# Clean nulls (optional depending on analysis)
new_customer = new_customer.dropna(subset=['first_name', 'gender'])
```

```python
# Merge customer demographic and address
customer_info = pd.merge(customer_demo, customer_address, how='inner', on='customer_id')

# Merge with transactions
full_data = pd.merge(transactions, customer_info, how='inner', on='customer_id')

full_data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Age distribution
sns.histplot(customer_demo['age'], bins=20, kde=True)
plt.title("Customer Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

# Gender split
sns.countplot(x='gender', data=customer_demo)
plt.title("Gender Distribution")
plt.show()
```

```python
# Top selling products
top_products = full_data['product_id'].value_counts().head(10)
top_products.plot(kind='bar', title='Top 10 Products by Purchase Count')
plt.xlabel('Product ID')
plt.ylabel('Number of Purchases')
plt.show()
```

```python
#online vs in store sales
sns.countplot(x='online_order', data=full_data)
plt.title("Online vs In-Store Orders")
plt.xticks([0, 1], ['In Store', 'Online'])
plt.ylabel("Orders")
plt.show()
```

```python
# Export cleaned and merged data
full_data.to_csv("kpmg_cleaned_data.csv", index=False)
```

```python
# Load the full CSV
df = pd.read_csv('kpmg_cleaned_data.csv')

# Check column names to decide how to split
print(df.columns)

# Example column patterns — update these based on your actual column names:
customer_demo_cols = ['customer_id', 'gender', 'DOB', 'job_industry_category', 'wealth_segment', 'owns_car', 'tenure']
customer_address_cols = ['customer_id', 'address', 'postcode', 'state', 'country', 'property_valuation']
transactions_cols = ['transaction_id', 'product_id', 'customer_id', 'transaction_date', 'online_order', 'order_status', 'list_price', 'standard_cost', 'brand', 'product_line', 'product_class', 'product_size']
customer_transaction_cols = ['customer_id', 'transaction_id']

# Create separate DataFrames
customer_demo = df[customer_demo_cols].drop_duplicates()
customer_address = df[customer_address_cols].drop_duplicates()
transactions = df[transactions_cols].drop_duplicates()
customer_transaction = df[customer_transaction_cols].drop_duplicates()

# Export to separate CSVs
customer_demo.to_csv('CustomerDemographic.csv', index=False)
customer_address.to_csv('CustomerAddress.csv', index=False)
transactions.to_csv('Transactions.csv', index=False)
customer_transaction.to_csv('CustomerTransaction.csv', index=False)

print("Files saved successfully!")
```
